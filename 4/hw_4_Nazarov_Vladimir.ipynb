# №1
Нарисуйте графики, показывающий количество браков и разводов на душу населения в США за период с 2000 года по 2014. Отобразите легенду, подпишите оси.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# there are not 2012, 2013, 2014 rows in this file !!!!
us_marriage_divorce = pd.read_csv('data/us-marriages-divorces-1867-2014.csv').dropna().tail(12)

x = us_marriage_divorce['Year'].values
y1 = us_marriage_divorce['Marriages'].values
y2 = us_marriage_divorce['Divorces'].values

plt.figure()

plt.plot(x, y1, label='Marriages')
plt.plot(x, y2, label='Divorces')
plt.xlabel('Year')
plt.ylabel('Number')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
```

# №3
Постройте вертикальную столбчатую диаграмму, сравнивающую количество браков и разводов на душу населения в США в 1900, 1950, 2000 годах.

```python
us_marriage_divorce = pd.read_csv('data/us-marriages-divorces-1867-2014.csv')
years = np.array([1900, 1950, 2000])

category1_values = []
category2_values = []

for y in years:
    ind = y - us_marriage_divorce['Year'].iloc[0]
    category1_values.append(us_marriage_divorce['Marriages'].iloc[ind])
    category2_values.append(us_marriage_divorce['Divorces'].iloc[ind])

category1_values = np.array(category1_values)
category2_values = np.array(category2_values)

plt.figure()

plt.bar(years + 7,
        category1_values,
        color='blue',
        edgecolor='none',
        width=14,
        align='center',
        label='Marriages')

plt.bar(years - 7,
        category2_values,
        color='orange',
        edgecolor='none',
        width=14,
        align='center',
        label='Divorces')

plt.xlabel('Year')
plt.ylabel('Number')
plt.xticks(years, [str(year) for year in years])
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
```

#  №5
Постройте точечный график, показывающий зависимость между количеством выданных докторских степеней в Computer Science и общей выручкой от продаж компьютерных игр.

```python
arcade_revenue_vs_cs_doctorates = pd.read_csv('data/arcade-revenue-vs-cs-doctorates.csv')
beg_year = arcade_revenue_vs_cs_doctorates['Year'].iloc[0]
ind = arcade_revenue_vs_cs_doctorates['Year'].size-1
end_year = arcade_revenue_vs_cs_doctorates['Year'].iloc[ind]

x = np.arange(beg_year, end_year + 1)
y1 = arcade_revenue_vs_cs_doctorates['Total Arcade Revenue (billions)'] * 1000
y2 = arcade_revenue_vs_cs_doctorates['Computer Science Doctorates Awarded (US)']

plt.figure()
plt.scatter(x, y1, c='red', marker='s', edgecolor='blue', label='Total Arcade Revenue (millions)')
plt.scatter(x, y2, c='red', marker='^', edgecolor='green', label='Computer Science Doctorates Awarded (US)')
plt.xlabel('Year')
plt.ylabel('Number')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

# №6
Нарисуйте круговую диаграмму, показывающую соотношение способов умереть среди римских императоров.

```python
roman_emperors = pd.read_csv('data/roman-emperor-reigns.csv')
causes = roman_emperors['Cause_of_Death'].value_counts()

plt.figure(figsize=(4, 4))

plt.pie(causes, labels=causes.values.tolist(), startangle=90)
plt.legend(labels=causes.index.tolist(), bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
